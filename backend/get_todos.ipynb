```python
import boto3
import pandas as pd
# iterate dataframe create folder from each row
import os


# Create a DynamoDB client
dynamodb = boto3.resource('dynamodb')
table = dynamodb.Table('todo_main')

# Use the DynamoDB client to scan all items in the table
response = table.scan()

# Extract the items from the response
items = response['Items']

folder_base = "./exported_data"

# Convert the items to a pandas dataframe
df = pd.DataFrame(data=items)
df.head()
```

```python

# create folder for each row
for index, row in df.iterrows():
    folder = folder_base + "/" + row['group'] + '/' + row['name']
    print( "{} {}".format( row["id"], folder) )
    os.makedirs(folder , exist_ok=True)
    
```

```python
# Create a DynamoDB client
from boto3.dynamodb.conditions import Key, Attr

table_todo = dynamodb.Table('todo')

def fetchTodosByMainId( main_id ):
    response = table_todo.query(
        # by index 
        IndexName='listid_idx',
        KeyConditionExpression=Key('listid').eq(main_id)
    )
    items = response['Items']

    todos = pd.DataFrame(data=items)
    return todos


fetchTodosByMainId("_1673958806935")
```
